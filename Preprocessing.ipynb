Credit Card Fraud Detection Dataset

```python
df_fraud = pd.read_csv("Fraud_Detection_Data_Train_sampled.csv")
```

```python
df_fraud.head()
```

```python
df_fraud.info()
```

```python
df_fraud.describe()
```

```python
df_fraud_dups = df_fraud[df_fraud.duplicated()]
print(df_fraud_dups)
```

```python
null_rows = df_fraud[df_fraud.isnull().any(axis=1)]
print(null_rows)
```

```python
df_fraud.rename(columns={"Unnamed: 0": "id"}, inplace=True)
```

```python
df_fraud["id"] = df_fraud["id"].astype(str)
```

```python
df_fraud.head()
```

```python
df_fraud.to_csv("Fraud_Detection_Data_Train_sampled.csv", index=False)
```
